#  Data Cleaning & Manipulation

```python
# Here I will check the Null Values present in the Columns

df.isnull().sum().sort_values(ascending = False)
```

# Filling the Null Values Present in the Columns

```python
df['company']=df['company'].fillna(0)
df['agent']=df['agent'].fillna(df['agent'].median())
df['country']=df['country'].fillna(df['country'].mode()[0])
df['children']=df['children'].fillna(df['children'].median())


df.isnull().sum()
```

# Handling the Duplicate Values

```python
df.duplicated().sum()
```

```python
# I found the 32020 Duplicate Values and So I will drop it now and then check the shape of the Dataset

df = df.drop_duplicates()

df.shape
```

```python
df.duplicated().sum()
```

```python
df.head()
```

# Feature Engineering (I created two new Columns here)

```python
# 1st Column = "Total_People" = from children, adults & babies column
# 2nd Column = "Total_Stay" = from stays_in_weekend_nights & stays_in_week_nights

df["Total_People"] = df["children"]+df["adults"]+df["babies"]

df["Total_Stay"] = df["stays_in_weekend_nights"]+df["stays_in_week_nights"]
```
